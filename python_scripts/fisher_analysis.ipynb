```python
# Useful functions for fisher analysis

def eigsorted(cov):
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    return vals[order], vecs[:,order]

def get_fisher(derivs, cov, hartlap):
    # Invert covariance
    cov_inv = np.linalg.inv(cov)*hartlap

    # Check if inverse covariance is symmetric
    is_sym=np.allclose(cov_inv, cov_inv.T)
    if not is_sym:
        print("Caution: inverse covariance not symmetric!")

    # Check if inverse covariance is actual inverse
    is_inv=np.allclose(np.eye(cov_inv.shape[0]), cov.dot(cov_inv/hartlap))
    if not is_inv:
        print("Caution: inverse covariance times covariance is not unity!")
            
            
    # Calculate fisher
    fisher_np=np.dot(derivs, cov_inv).dot(derivs.T)


    # Check if fisher matrix is symmetric
    is_sym=np.allclose(fisher_np, fisher_np.T)
    if not is_sym:
        print("Caution: fisher matrix not symmetric!")

    return fisher_np


def plotFisherEllipse(ax, i,  j, fisher_inv, param1, param2, color='k', fill=False, alpha=1.0):
    cov=np.array([[fisher_inv[i,i], fisher_inv[i,j]],[fisher_inv[j,i], fisher_inv[j,j]]])
 
    vals, vecs = eigsorted(cov)
    theta=np.degrees(np.arctan2(*vecs[:,0][::-1]))
    w, h = 2*np.sqrt(vals)

    ellipse=Ellipse(xy=(param1, param2), width=w, height=h, angle=theta, color=color, fill=fill, alpha=alpha)


    ax.add_patch(ellipse)
    ax.plot(param1, param2, marker='+', color='k')

def plotFisherProb(ax, i, fisher_inv, param, color='k', alpha=1.0):
    sigma=np.sqrt(np.diag(fisher_inv)[i])
    x=np.linspace(param-3*sigma, param+3*sigma)
    ax.plot(x, np.exp(-0.5*pow((x-param)/sigma,2))/np.sqrt(2*np.pi)/sigma, color=color, alpha=alpha)


def addFisherplots(axs, fisher, params, param_names, color='b', fill=False, alpha=1.0):
    fisher_inv=np.linalg.inv(fisher)
    Nparams=len(params)
    for i, param1 in enumerate(params):
        axs[Nparams-1, i].set_xlabel(param_names[i])

        for j, param2 in enumerate(params):
            axs[j, 0].set_ylabel(param_names[j])

            if i==j:
                plotFisherProb(axs[i,j], i, fisher_inv, param1, color=color, alpha=alpha)
            elif i>j:
                plotFisherEllipse(axs[i,j], j, i, fisher_inv, param2, param1, color=color, fill=fill, alpha=alpha)
            else:
                axs[i,j].set_axis_off()
```
