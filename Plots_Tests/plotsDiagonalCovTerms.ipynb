```python
#load data
cov_fft=np.loadtxt(dir_ms+"cov_MS_fft_sigma_0.00_n_1048576.00_thetaMax_1.87.dat")
cov_T1inf=np.loadtxt(dir_ms+"cov_infinite_term1Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")
cov_T2=np.loadtxt(dir_ms+"cov_square_term2Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")
cov_T4=np.loadtxt(dir_ms+"cov_infinite_term4Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")
cov_T5=np.loadtxt(dir_ms+"cov_infinite_term5Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")
cov_T6=np.loadtxt(dir_ms+"cov_square_term6Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")/(2*np.pi)
#cov_T7=np.loadtxt(dir_ms+"cov_infinite_term7Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")
cov_T7=np.loadtxt(dir_ms+"cov_infinite_term7Numerical_sigma_0.00_n_1048456.68_thetaMax_1.87_gpu.dat")

cov_T1inf_meas=np.loadtxt(dir_ti+"T_1_measured.dat")
cov_T2_meas=np.loadtxt(dir_ti+"T_2_measured.dat")
cov_T4_meas=np.loadtxt(dir_ti+"T_4_measured.dat")
cov_T5_meas=np.loadtxt(dir_ti+"T_5_measured.dat")
cov_T6_meas=np.loadtxt(dir_ti+"T_6_measured.dat")
cov_T7_meas=np.loadtxt(dir_ti+"T_7_measured.dat")

```

```python
fig, axs=plt.subplots(nrows=3, ncols=2, sharex=True, sharey=True, figsize=(15,15))
plt.subplots_adjust(hspace=0, wspace=0)
for ax in axs.ravel():
    ax.set_ylim((1e-22, 3e-15))
    ax.set_yscale('log')

axs[0,0].plot(np.diag(cov_T1inf), label="T1(model)", color='C0')
axs[0,0].plot(np.diag(cov_T1inf_meas), label="T1(measured)", ls='--', color='C0')
axs[0,1].plot(np.diag(cov_T2), label="T2(model)", color='C1')
axs[0,1].plot(np.diag(cov_T2_meas), label="T2(measured)", ls='--', color='C1')
axs[1,0].plot(np.diag(cov_T4), label="T4(model)", color='C2')
axs[1,0].plot(np.diag(cov_T4_meas), label="T4(measured)", ls='--', color='C2')
axs[1,1].plot(np.diag(cov_T5), label="T5(model)", color='C3')
axs[1,1].plot(np.diag(cov_T5_meas), label="T5(measured)", ls='--', color='C3')
axs[2,0].plot(np.diag(cov_T6), label="T6(model)", color='C4')
axs[2,0].plot(np.diag(cov_T6_meas), label="T6(measured)", ls='--', color='C4')
axs[2,1].plot(cov_T7[:,3], label="T7(model)", color='C5')
axs[2,1].plot(np.diag(cov_T7_meas), label="T7(measured)", ls='--', color='C5')


fig.legend()
```
